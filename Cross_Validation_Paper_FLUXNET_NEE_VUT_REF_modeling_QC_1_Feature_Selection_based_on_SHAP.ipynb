#Optuna MLP

```python
np.random.seed(23)
tf.random.set_seed(16)
tf.keras.utils.set_random_seed(2018)

callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                             min_delta=0.001,
                                             patience=5,
                                             verbose=0,
                                             mode='auto',
                                             restore_best_weights=True)

def objective(trial):
  # Sample hyperparameters
  n_units = trial.suggest_int('n_units', 1, 350)
  n_layers = trial.suggest_int('n_layers', 0, 2)
  learning_rate = trial.suggest_float('learning_rate', 1e-5, 1e-2, log=True)
  batch_size = trial.suggest_int('batch_size', 1, 150)
  n_epochs = trial.suggest_int('n_epochs', 5, 350)
  activation = trial.suggest_categorical('activation', ['relu', 'tanh', 'sigmoid'])
  kernel_initializer = trial.suggest_categorical('kernel_initializer', ['uniform', 'normal', 'he_normal', 'he_uniform', 'glorot_uniform', 'glorot_normal'])
  n_folds = trial.suggest_int('n_folds', 4, 8)

  # Build the MLP model with sampled hyperparameters
  model = Sequential()

  # Add input layer
  model.add(Dense(units=n_units, kernel_initializer=kernel_initializer, activation=activation))

  # Add hidden layers and allow different hyperparameters values for each layer
  for i in range(n_layers):
    layer_units = trial.suggest_int(f'n_units_{i}', 1, 250)
    layer_activation = trial.suggest_categorical(f'activation_{i}', ['relu', 'tanh', 'sigmoid'])
    layer_l2 = trial.suggest_float(f'l2_reg_strength_{i}', 0.001, 0.5, log=True)
    layer_kernel_initializer = trial.suggest_categorical(f'kernel_initializer_{i}', ['uniform', 'normal', 'he_normal', 'he_uniform', 'glorot_uniform', 'glorot_normal'])
    model.add(Dense(units=layer_units,
                    kernel_initializer=layer_kernel_initializer,
                    activation=layer_activation,
                    kernel_regularizer=regularizers.l2(layer_l2)))

  # Add output layer
  model.add(Dense(1, activation='linear'))

  # Compile the model
  model.compile(optimizer=Adam(learning_rate=learning_rate), loss='mean_squared_error')

  # K-Fold Cross-Validation within objective function
  kfold = KFold(n_splits=n_folds, shuffle=True, random_state = 27)
  scores = []
  for train_index, val_index in kfold.split(X):
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    Y_train, Y_val = Y.iloc[train_index], Y.iloc[val_index]

    # Train the model
    model.fit(X_train, Y_train, epochs=n_epochs, batch_size=batch_size, verbose=0,
              validation_data=(X_val, Y_val),
              callbacks=[callback])

    # Predict on the validation set
    Y_pred_val = model.predict(X_val)

    # Calculate the mean squared error
    mse = mean_squared_error(Y_val, Y_pred_val)
    #print(f"Mean MSE: {mse}")
    scores.append(mse)

# Return the average MSE across folds
  #print(f"Average MSE across all folds: {np.mean(scores)}")
  return np.mean(scores)

# Create an Optuna study object
study = optuna.create_study(direction='minimize', sampler=optuna.samplers.TPESampler(seed=100))

# Run the optimization
study.optimize(objective, n_trials=200)

# Access the trials as a pandas DataFrame
trials_df = study.trials_dataframe()

# Print the best trial
best_trial = study.best_trial
print("Best Trial:")
print("Value: ", best_trial.value)
print("Params: ")

for key, value in best_trial.params.items():
  print(f"{key}: {value}")
```

```python
np.random.seed(23)
tf.random.set_seed(16)
tf.keras.utils.set_random_seed(2018)

callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                             min_delta=0.001,
                                             patience=5,
                                             verbose=0,
                                             mode='auto',
                                             restore_best_weights=True)

def ann():
  model = Sequential()
  model.add(Dense(units=29, kernel_initializer='he_normal', activation='relu'))
  model.add(Dense(units=211, kernel_initializer='he_uniform', activation='relu', kernel_regularizer=regularizers.l2(0.003843889888300198)))
  model.add(Dense(1, activation='linear'))

  model.compile(optimizer=Adam(learning_rate=0.00239556364161576), loss='mean_squared_error')
  return model

# Define the K-fold cross-validation parameters
kfold = KFold(n_splits=8, shuffle=True, random_state=27)
r2_scores = []
mse_scores = []

# Perform K-fold cross-validation
for train_index, val_index in kfold.split(X):
  X_train, X_val = X.iloc[train_index], X.iloc[val_index]
  Y_train, Y_val = Y.iloc[train_index], Y.iloc[val_index]

  # Create a new model instance for each fold to avoid overfitting
  model = ann()

  # Train the model
  model.fit(X_train, Y_train, validation_data = (X_val, Y_val), epochs=332, batch_size=48, callbacks = [callback], verbose=0)

  # Evaluate the model
  Y_pred = model.predict(X_val)
  r2 = r2_score(Y_val, Y_pred)
  mse = mean_squared_error(Y_val, Y_pred)
  r2_scores.append(r2)
  mse_scores.append(mse)

# Print the R2 score results
print(("R2_Score Mean & SD: %.3f%% (%.3f%%)" % (np.mean(r2_scores), np.std(r2_scores))),
      ("MSE_Score Mean & SD: %.3f (+/- %.3f)" % (np.mean(mse_scores), np.std(mse_scores))))
```

# Optuna Random Forest

```python
def objective(trial):
    n_estimators = trial.suggest_int('n_estimators', 100, 500)
    max_depth = trial.suggest_int('max_depth', 1, 5)
    min_samples_split = trial.suggest_int('min_samples_split', 5, 150)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 5, 150)
    max_features = trial.suggest_int('max_features', low=1, high=4)
    n_folds = trial.suggest_int('n_folds', 4, 8)

    model = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        max_features=max_features,
        random_state=23
    )

    # K-Fold Cross-Validation within objective function
    kfold = KFold(n_splits=n_folds, shuffle=True, random_state=27)
    scores = []
    for train_index, val_index in kfold.split(X):
        X_train, X_val = X.iloc[train_index], X.iloc[val_index]
        Y_train, Y_val = Y.iloc[train_index], Y.iloc[val_index]

        # Train the model
        model.fit(X_train, Y_train)

        # Predict on the validation set
        Y_pred_val = model.predict(X_val)

        # Calculate the mean squared error
        mse = mean_squared_error(Y_val, Y_pred_val)
        # print(f"Mean MSE: {mse}")
        scores.append(mse)

    # Return the average MSE across folds
    #print(f"Average MSE across all folds: {np.mean(scores)}")
    return np.mean(scores)

# Create an Optuna study object
study = optuna.create_study(direction='minimize', sampler=optuna.samplers.TPESampler(seed=100))

# Run the optimization
study.optimize(objective, n_trials=200)

# Access the trials as a pandas DataFrame
trials_df = study.trials_dataframe()

# Print the best trial
best_trial = study.best_trial
print("Best Trial:")
print("Value: ", best_trial.value)
print("Params: ")

for key, value in best_trial.params.items():
    print(f"{key}: {value}")
```
